A notebook to train salmon component detection models

```python
from ultralytics import YOLO
import torch
import shutil
import sys
import os
sys.path.append(os.path.split(os.getcwd())[0] + '\\format_dataset')
from format_dataset import generate_yaml_file
```

```python
# bounding_box_detection or keybox_detection
task = 'keybox_detection'

# Path
YOLO_root = 'C:\\Users\\espen\\Documents\\work\\PhD\\papers\\salmon_tracking\\data\\YOLO_data\\'
generate_yaml_file(YOLO_root, task, ['head', 'dorsal_fin', 'adi_fin', 'tail_fin', 'anal_fin','pelv_fin','pec_fin', 'body'], additional_kps=[])
```

```python
## Use GPU if accessible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device: ", device, f"({torch.cuda.get_device_name(device)})" if torch.cuda.is_available() else "")
```

```python
if task.startswith('keybox_detection'):
    model = YOLO('yolov8n-pose')
elif task == 'bounding_box_detection':
    model = YOLO('yolov8n')

model.train(data=YOLO_root + '\\YOLO_data\\' + 'config.yml', 
            epochs=10000, 
            imgsz=640, 
            workers = 1,
            batch = 8,  
            project = YOLO_root + '\\YOLO_models\\' + task, 
            patience = 10000,
            degrees = 20,
            scale = 0.8,
            perspective = 0.0001,
           )
```
